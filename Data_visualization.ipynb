```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

A line chart is a type of graph used to show data changes over time. It connects data points with lines, making it easy to understand trends.

Matplotlib is a popular Python library used for data visualization. It helps us create graphs like line charts, bar graphs, and more.

```python
days=[1,2,3,4,5,6]
temp=[31,28,34,39,40,32]
```

```python
#line chart
```

```python
plt.plot(days,temp)
```

```python
# plot multiple lines in the same figure
days=[1,2,3,4,5,6]
max_temp=[31,28,34,39,40,32]
min_temp=[20,21,19,22,30,20]
```

```python
plt.plot(days,max_temp)
plt.plot(days,min_temp)
plt.show()
```

```python
#changing the color of the lines
plt.plot(days,max_temp,color='cyan')
plt.plot(days,min_temp,color='blue')
plt.show()
```

```python
#label the graph
plt.plot(days,max_temp,color='cyan')
plt.plot(days,min_temp,color='blue')
plt.xlabel("Days")
plt.ylabel("Temerature recorded")
plt.title("Days vs the temp")
plt.show()
```

```python
#to add marker
plt.plot(days,max_temp,color='cyan',marker="*",markersize=10,linestyle="--",markerfacecolor='yellow')
plt.plot(days,min_temp,color='blue',marker="o",markersize=12,linestyle=":")
plt.xlabel("Days")
plt.ylabel("Temerature recorded")
plt.title("Days vs the temp")
plt.show()
```

```python
#how to put legends
plt.plot(days,max_temp,color='cyan',marker="*",markersize=10,linestyle="--",markerfacecolor='yellow')
plt.plot(days,min_temp,color='blue',marker="o",markersize=12,linestyle=":")
plt.xlabel("Days")
plt.ylabel("Temerature recorded")
plt.title("Days vs the temp")
plt.legend(["max_temp","Min_temp"],loc="lower left")
plt.show()
```

```python
#provide grid to the graph
#sns.set_style("darkgrid")
#sns.set_style("whitegrid")
plt.figure(figsize=(8,6))
plt.plot(days,max_temp,color='cyan',marker="*",markersize=10,linestyle="--",markerfacecolor='yellow')
plt.plot(days,min_temp,color='blue',marker="o",markersize=12,linestyle=":")
plt.xlabel("Days")
# Show grid (optional but nice)
plt.grid()
plt.ylabel("Temerature recorded")
plt.title("Days vs the temp")
plt.legend(["max_temp","Min_temp"],loc="lower left")
plt.show()
```

```python
# create a line chart for the fora university representing the last 7 year
# passing score and the no. student get placed every year
passing_score=[98,95,95,90,80,87,99]
year=[2017,2018,2019,2020,2021,2022,2023]
placed=[78,80,100,70,75,77,90]
```

```python
#STACK THE BAR chart
```

```python
list1=['item1','item2','item3','item4']
list2=[25,40,25,30]
list3=[20,30,20,25]
plt.bar(list1,list2,color='red')
plt.bar(list1,list3,color='orange')
plt.legend([list2,list3])
plt.show()
```

```python
#multi stack bar
x=['item1','item2','item3','item4']
y1=np.array([10,20,15,30])
y2=np.array([20,25,20,25])
y3=np.array([12,15,19,6])
y4=np.array([10,29,15,19])
plt.bar(x,y1,color='red')
plt.bar(x,y2,bottom=y1,color='orange')
plt.bar(x,y3,bottom=y1+y2,color='blue')
plt.bar(x,y4,bottom=y1+y2+y3,color='green')
plt.legend(['a','b','c','d'])
plt.show()
```
