```python
import pandas as pd
```

```python
df = pd.read_csv('car.csv')
```

```python
df.head()
```

```python
# Most expensive car for each car company make.
df.groupby('Make')['MSRP'].max()
```

```python
#Most expensive convertible vehicle for each company.
df[df['Vehicle Style'] == 'Convertible'].groupby('Make')['MSRP'].max()
```

```python
#List all the cars which are automatic.
df[df['Transmission Type'] == 'AUTOMATIC']
```

```python
#premium cars of each car company
premium = df[ (df['Engine Fuel Type'] == 'premium unleaded (required)') | (df['Engine Fuel Type'] == 'premium unleaded (recommended)') ]
premium.groupby('Make')['Engine Fuel Type'].count()
```

```python
#Average price of cars each car company make.
df.groupby('Make')['MSRP'].mean()
```
